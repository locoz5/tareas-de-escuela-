```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar los datos
datos = pd.read_csv('fifa_eda.csv')

# 2. Mostrar datos
print("Primeras filas:")
print(datos.head(3))  # Muestra solo 3 filas para simplificar

# 3. Estadísticas básicas
print("\nEstadísticas:")
print(datos.describe())

# 4. Convertir columnas a números
datos['Value'] = pd.to_numeric(datos['Value'].str.replace('€', '').str.replace('M', ''), errors='coerce')
datos['Wage'] = pd.to_numeric(datos['Wage'].str.replace('€', '').str.replace('K', ''), errors='coerce')

# 5. Años jugando (simplificado)
datos['Years Playing'] = 2023 - datos['Joined']  # Usamos 2023 como año actual

# 6. Jugadores mexicanos
mexicanos = datos[datos['Nationality'] == 'Mexico']
print("\nJugadores de México:")
print(mexicanos[['Name', 'Age']].head())  # Solo nombre y edad

# 7. Ordenar por sueldo
datos_ordenados = datos.sort_values('Wage', ascending=False)
print("\nTop sueldos:")
print(datos_ordenados[['Name', 'Wage']].head(3))

# 8. Gráfico simple (opcional)
datos['Joined'].value_counts().sort_index().plot(kind='bar', figsize=(10,4))
plt.title('Jugadores por año')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar los datos con manejo de errores
try:
    datos = pd.read_csv('fifa_eda.csv', encoding='utf-8')
    print("Datos cargados correctamente")
except FileNotFoundError:
    print("Error: Archivo no encontrado")
    exit()
except Exception as e:
    print(f"Error al cargar archivo: {str(e)}")
    exit()

# 2. Mostrar datos con formato mejorado
print("\n=== Primeras 3 filas ===")
display(datos.head(3))

# 3. Estadísticas mejoradas
print("\n=== Estadísticas Descriptivas ===")
display(datos.describe(include='all'))  # Incluye todas las columnas

# 4. Conversión numérica mejorada
def clean_currency(value):
    if isinstance(value, str):
        return float(value.replace('€', '').replace('M', 'e6').replace('K', 'e3'))
    return value

try:
    datos['Value'] = datos['Value'].apply(clean_currency)
    datos['Wage'] = datos['Wage'].apply(clean_currency)
    print("\nColumnas convertidas a numéricas")
except Exception as e:
    print(f"\nError en conversión numérica: {str(e)}")

# 5. Cálculo de años jugando con validación
current_year = pd.Timestamp.now().year
if 'Joined' in datos.columns:
    datos['Years Playing'] = current_year - datos['Joined']
    print("\nColumna 'Years Playing' creada")
else:
    print("\nAdvertencia: Columna 'Joined' no encontrada")

# 6. Jugadores mexicanos con más información
if 'Nationality' in datos.columns:
    mexicanos = datos[datos['Nationality'] == 'Mexico']
    print("\n=== Jugadores Mexicanos ===")
    display(mexicanos[['Name', 'Age', 'Club', 'Overall']].head())
else:
    print("\nAdvertencia: Columna 'Nationality' no encontrada")

# 7. Ordenamiento por salario con verificación
if 'Wage' in datos.columns:
    datos_ordenados = datos.sort_values('Wage', ascending=False)
    print("\n=== Top 3 Salarios ===")
    display(datos_ordenados[['Name', 'Nationality', 'Wage']].head(3))
else:
    print("\nAdvertencia: Columna 'Wage' no encontrada")

# 8. Gráfico mejorado
plt.figure(figsize=(12, 6))
if 'Joined' in datos.columns:
    datos['Joined'].value_counts().sort_index().plot(
        kind='bar', 
        color='skyblue',
        edgecolor='black'
    )
    plt.title('Distribución de Jugadores por Año de Ingreso', fontsize=14)
    plt.xlabel('Año', fontsize=12)
    plt.ylabel('Número de Jugadores', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
else:
    print("\nAdvertencia: No se puede generar gráfico - Columna 'Joined' no encontrada")
```
