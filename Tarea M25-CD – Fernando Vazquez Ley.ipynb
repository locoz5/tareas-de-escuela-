```python

# Temática: Razas de perros y sus características


# ### 1. Función lambda para raíz cuadrada
# **Ejemplo:** Calcular la raíz cuadrada del peso de un Golden Retriever (32 kg)

# %%
# Sin lambda
import math

def raiz_cuadrada(n):
    return math.sqrt(n)

print("Sin lambda:", raiz_cuadrada(32))

# Con lambda
raiz_lambda = lambda x: math.sqrt(x)
print("Con lambda:", raiz_lambda(32))

# %% [markdown]
# ### 2. Map para obtener largo de palabras
# **Ejemplo:** "GranDanés Caniche PastorAleman" → [9, 7, 12]

# %%
# Sin map
cadena_perros = "GranDanés Caniche PastorAleman"
palabras = cadena_perros.split()
longitudes = []

for palabra in palabras:
    longitudes.append(len(palabra))
    
print("Sin map:", longitudes)

# Con map
longitudes_map = list(map(len, palabras))
print("Con map:", longitudes_map)

# %% [markdown]
# ### 3. Reduce para calcular producto
# **Ejemplo:** Producto de cachorros por camada [3, 2, 4] = 24

# %%
# Sin reduce
lista_cachorros = [3, 2, 4]
producto = 1

for num in lista_cachorros:
    producto *= num
    
print("Sin reduce:", producto)

# Con reduce
from functools import reduce

producto_reduce = reduce(lambda a, b: a * b, lista_cachorros)
print("Con reduce:", producto_reduce)

# %% [markdown]
# ### 4. Filter para palabras con mayúsculas/números
# **Ejemplo:** ["Labrador2", "boxer", "GoldenRetriever"] → ["Labrador2", "GoldenRetriever"]

# %%
# Sin filter
lista_razas = ["Labrador2", "boxer", "GoldenRetriever"]
resultado = []

for palabra in lista_razas:
    tiene_especial = False
    for letra in palabra:
        if letra.isupper() or letra.isdigit():
            tiene_especial = True
            break
    if tiene_especial:
        resultado.append(palabra)
        
print("Sin filter:", resultado)

# Con filter
resultado_filter = list(filter(lambda palabra: any(c.isupper() or c.isdigit() for c in palabra), lista_razas))
print("Con filter:", resultado_filter)
```
